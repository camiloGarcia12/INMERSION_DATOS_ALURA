## **Para darle un valor y utilizarlo solo para la información que quiero ver**

# **COMO EL PROGRAMA ESTA TOMANDO EL PUNTO COMO UN SEPARADOR SE VA A ELIMINAR EL . CON REPLACE**




```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','', regex = True)
```

```python
inmuebles[['Precio', 'Barrio']]
```

```python
inmuebles.info()
```

# LOS DATOS SIGUEN SIENDO OBJETOS, POR LO TANTO SE CONVERTIRA EN FLOAT (es un tipo númerico que permite representar un numero positivo y negativo y asi convierte datos a numeros reales) Y ADICIONAL SE MODIFICA EL PRECIO RESTANDOLES 6 CEROS POR SU ALTO NUMERO

```python
inmuebles['Precio_Milllon'] = inmuebles.Precio.astype('float')/1000000

inmuebles.info()
```

## **UTILIZACION DE GRAFICAS MAS DETALLADAS**

 



```python
import matplotlib.pyplot as plt
import seaborn as sns
#sns = para manejar los datos

plt.figure(figsize=(10,8)) #para presentar la imagen y sus dimenciones
hue_order = ['Casa', 'Apartamento', 'Finca'] #Hue_order = para ordenas la variables de interes.
palette = ['red', 'blue', 'black'] # Aquí se definen los colores correspondientes a cada tipo de inmueble
grafica = sns.histplot(data=inmuebles, x='Precio_Milllon', kde=True, hue='Tipo', hue_order=hue_order, palette=palette)  #hue=La variable que queremos en la grafica, otra dimencion por asi decirlo.
#kde =  kernel density estimation

grafica.set_title('Distribución de valores de los inmuebles de Bógota')
plt.xlim((50,1000))
plt.xlabel('Precio (millones de pesos)')
plt.ylabel('Frecuencia')
plt.savefig('/content/drive/MyDrive/BASE_DATOS/BASE_DATOS_INMUEBLES_AULA_1.png',format='png')
plt.show()

#plt.xlabel e ylabel = sirven para nombrar los labels.
#plt.xlim = son las lineas o limite qque queremos para el eje x
#plt.xlim = son las lineas o limite qque queremos para el eje y
#plt.savefig = es para salvar o exportar la imagen
```

```python
inmuebles['Precio'] = pd.to_numeric(inmuebles['Precio'])
inmuebles['Area'] = pd.to_numeric(inmuebles['Area'])
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

inmuebles['Precio_m2'] = inmuebles['Precio'] / inmuebles['Area']

plt.figure(figsize=(40,20))
sns.scatterplot(x=inmuebles['Barrio'], y=inmuebles['Precio_m2'], alpha=0.5)
sns.kdeplot(x=inmuebles['Barrio'], y=inmuebles['Precio_m2'], cmap='Reds', shade=True)
plt.xticks(rotation=90)
plt.xlabel('Barrio')
plt.ylabel('Precio por metro cuadrado (pesos)')
plt.title('Relación entre precio por metro cuadrado y barrio de los inmuebles')
plt.savefig('/content/drive/MyDrive/BASE_DATOS/BASE_DATOS_INMUEBLES_AULA_2_DESAFIO.png',format='png')
plt.show()

#1e7 significa 1 x 10^7, es decir, 10,000,000.
```
